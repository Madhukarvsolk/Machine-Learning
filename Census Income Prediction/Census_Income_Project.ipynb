# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
lr=LinearRegression()   #making an object of linear regression class
```

```python
#As per the question independent variable is “education.num”.
x=data[['education.num']]
#dependent variable is “hours.per.week”
y=data['hours.per.week']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=1)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
y_pred   #These are the predicted values
```

```python
y_test #these are actual values
```

```python
error=y_test-y_pred
error
```

```python
print('mean_squared_error :',mean_squared_error(y_test,y_pred))

print('root-mean-square error :',np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
data=data.apply(L.fit_transform) #L was object of label encoder
data.head()
```

```python
#independent
x=data.iloc[:,:-1]
#dependent
y=data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=1)
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)

print('confusion_matrix :')
print(confusion_matrix(y_pred,y_test))
print('accuracy_score :',accuracy_score(y_test,y_pred))
```
